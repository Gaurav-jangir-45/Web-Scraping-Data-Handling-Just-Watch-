````python
#import all necessary labraries
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import numpy as np
````

## **Scrapping Movies Data**

````python
def fetch_movie_urls(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'
    }
    response = requests.get(url, headers=headers)
    if response.status_code != 200:
        return "Failed to retrieve the page, status code:", response.status_code
    soup = BeautifulSoup(response.text, 'html.parser')
    return soup


url = 'https://www.justwatch.com/in/movies?release_year_from=2000'
soup=fetch_movie_urls(url)
print(soup.prettify())

## Hint : Use the following code to extract the film urls
# movie_links = soup.find_all('a', href=True)
# movie_urls = [link['href'] for link in movie_links if '/movie/' in link['href']]

# url_list=[]
# for x in movie_urls:
#   url_list.append('https://www.justwatch.com'+x)
````

## **Fetching Movie URL's**

````python
# Write Your Code here
# Extrecting movie objects
movie_links = soup.find_all('a', href=True)

# Extracting all movie links
movie_urls = [link['href'] for link in movie_links if '/movie/' in link['href']]

url_list=[] # Creating all movie links
for x in movie_urls:
  url_list.append('https://www.justwatch.com'+x)
# print movie_list
print(url_list)

````

## **Scrapping Movie Title**

````python
# Write Your Code here
movie_object_list = [] # Creating a empty list to store movie objects
for url in url_list:
  # Getting response for all the movie urls
  soup_1 = fetch_movie_urls(url)
  movie_object_list.append(soup_1)
````

````python
# Write Your Code here
movie_titles =[] # Empty list to store all movie titles

# Looping through the movie objects list
for soup_1 in movie_object_list:
  # Extracting all the movie titles from soup
  titles = soup_1.find_all('h1', class_="title-detail-hero__details__title")
# removing all extra texts
  for title in titles:
    movie_titles.append(re.sub(r'\(\d{4}\)', '', title.text.strip()))
# print all the mvoie titles
print(movie_titles)
````

## **Scrapping release Year**

````python
# Write Your Code here
release_years = [] # Creating a empty list for storing release years

# Extracting the release years from movie_object_list
for soup_1 in movie_object_list:
  release_year = soup_1.find_all('span',class_ = "release-year")
  for year in release_year:
    # Removing parentheses form movie release years
    release_years.append(year.text.replace('(','').replace(')','') .strip())
# print release years
print(release_years)
````

## **Scrapping Genres**

````python
# Write Your Code here
movie_genre_list =[]
for soup_1 in movie_object_list:
      # Selecting only those h3 whose heading is genres
    h3_tag = soup_1.find('h3', class_='detail-infos__subheading', string='Genres')

    if h3_tag: # Checking if h3_tag exists
        # Check if the next sibling is a div with class "detail-infos__value"
        div_tag = h3_tag.find_next_sibling('div', class_='detail-infos__value')

        if div_tag: # checking if div_tag exists
            movie_genre_list.append(div_tag.text.strip())
        else:
            movie_genre_list.append("Genre Not Listed")
    else:
         movie_genre_list.append("Genre Not Listed")
# print movie_genre_list
print(movie_genre_list)


````

## **Scrapping IMBD Rating**

````python
# Write Your Code here
movie_rating_list =[] # Creating a empty list for storing movie ratings
# Extracting movie objects
for soup_1 in movie_object_list:
  # Extracting div_tag from movie objects
  div_tag = soup_1.find_all('div', class_="title-detail-hero-details__item")
  if div_tag: # Checking if div_tag exists
    for div in div_tag:
      span_tag = div.find('span', class_="imdb-score") # Extracting span_tag form div_tag
      if span_tag:# Checking if span_tag exists
        movie_rating_list.append(span_tag.text.strip())

  else:
    movie_rating_list.append("Rating Not Listed")
# Removig extra char from the rating list
movie_rating_list = [re.sub(r'\s*\(.*?\)', '', rating) for rating in movie_rating_list]
movie_rating_list = movie_rating_list + [np.nan]*(110-len(movie_rating_list))
# print movie_rating_list
print(movie_rating_list)
````

## **Scrapping Runtime/Duration**

````python
# Write Your Code here
movie_runtime_list =[]
for soup_1 in movie_object_list:
      # Selecting only those h3 whose heading is genres
    h3_tag = soup_1.find('h3', class_='detail-infos__subheading', string='Runtime')

    if h3_tag: # Checking if h3_tag exists
        # Check if the next sibling is a div with class "detail-infos__value"
        div_tag = h3_tag.find_next_sibling('div', class_='detail-infos__value')

        if div_tag: # checking if div_tag exists
            movie_runtime_list.append(div_tag.text.strip())
        else:
            movie_runtime_list.append(np.nan)
    else:
         movie_runtime_list.append(np.nan)
# print movie_runtime_list
print(movie_runtime_list)
````

## **Scrapping Age Rating**

````python
# Write Your Code here
movie_age_rating_list =[]
for soup_1 in movie_object_list:
      # Selecting only those h3 whose heading is genres
    h3_tag = soup_1.find('h3', class_='detail-infos__subheading', string='Age rating')

    if h3_tag: # Checking if h3_tag exists
        # Check if the next sibling is a div with class "detail-infos__value"
        div_tag = h3_tag.find_next_sibling('div', class_='detail-infos__value')

        if div_tag: # checking if div_tag exists
            movie_age_rating_list.append(div_tag.text.strip())
        else:
            movie_age_rating_list.append("Age Rating Not Listed")
    else:
         movie_age_rating_list.append("Age Rating Not Listed")
# print movie_age_rating_list
print(movie_age_rating_list)
````

## **Fetching Production Countries Details**

````python
# Write Your Code here
movie_country_list =[]
for soup_1 in movie_object_list:
      # Selecting only those h3 whose heading is genres
    h3_tag = soup_1.find('h3', class_='detail-infos__subheading', string=' Production country ')

    if h3_tag: # Checking if h3_tag exists
        # Check if the next sibling is a div with class "detail-infos__value"
        div_tag = h3_tag.find_next_sibling('div', class_='detail-infos__value')

        if div_tag: # checking if div_tag exists
            movie_country_list.append(div_tag.text.strip())
        else:
            movie_country_list.append(np.nan)
    else:
         movie_country_list.append(np.nan)
# print movie_country_list
print(movie_country_list)
````

## **Fetching Streaming Service Details**

````python
# Write Your Code here
movie_streaming_list =[] # Creating a empty list

for soup_1 in movie_object_list: # Extracting movie objects from movie_object_list

# Finding div tag with class = "picture-wrapper"
  div_tag = soup_1.find('div', class_="picture-wrapper")

  if div_tag: # Checking if div_tag exists

     # Inside div_tag finding picture_tag
      picture = div_tag.find('picture')

      if picture: # Checking if picture exists

        # Inside picture_tag finding img tag
        img = picture.find('img')

        if img: # Checking if img exists

          # Extracting alt text from img tag
          alt = img.get('alt')
          if alt: # Checking if alt exists
            movie_streaming_list.append(alt)
          else:
            movie_streaming_list.append("Streaming Service Not Listed")
        else:
          movie_streaming_list.append("Streaming Service Not Listed")
      else:
        movie_streaming_list.append("Streaming Service Not Listed")
  else:
    movie_streaming_list.append("Streaming Service Not Listed")
# print movie_streaming_list
print(movie_streaming_list)
````

## **Task 2 :- Data Filtering & Analysis**

````python
# Write Your Code here
from datetime import datetime

# Convert 'Release Years' and 'Show Release Year' columns to datetime format
movie_df['Release Years'] = pd.to_datetime(movie_df['Release Years'], errors='coerce')
tv_show_df['Show Release Year'] = pd.to_datetime(tv_show_df['Show Release Year'], errors='coerce')

# Extract only the year from 'Release Years' and 'Show Release Year'
movie_df['Release Years'] = movie_df['Release Years'].dt.year
tv_show_df['Show Release Year'] = tv_show_df['Show Release Year'].dt.year

# Current year and two years ago year for filtering
current_year = datetime.now().year
two_years_ago_year = current_year - 2

# Convert 'IMDB Rating' column to numeric type
movie_df['IMDB Rating'] = pd.to_numeric(movie_df['IMDB Rating'], errors='coerce')
tv_show_df['IMDB Show Rating'] = pd.to_numeric(tv_show_df['IMDB Show Rating'], errors='coerce')

# Filter movies released in the last 2 years (2022, 2023, 2024) and having IMDb rating of 7 or higher
filtered_movies = movie_df[
    (movie_df['Release Years'].isin([2022, 2023, 2024])) &
    (movie_df['IMDB Rating'] >= 7)
]

# Filter TV shows released in the last 2 years (2022, 2023, 2024) and having IMDb rating of 7 or higher
filtered_tv_shows = tv_show_df[
    (tv_show_df['Show Release Year'].isin([2022, 2023, 2024])) &
    (tv_show_df['IMDB Show Rating'] >= 7)
]

# Reset index for the filtered DataFrames
filtered_movies = filtered_movies.reset_index(drop=True)
filtered_tv_shows = filtered_tv_shows.reset_index(drop=True)


````

## **Analyzing Top Genres**

````python
import pandas as pd
# Step 1: Split the 'Genre' and 'Show Genre' columns into lists of genres
filtered_movies['Genre'] = filtered_movies['Genre'].str.split(', ')
filtered_tv_shows['Show Genre'] = filtered_tv_shows['Show Genre'].str.split(', ')

# Step 2: Use .explode() to separate each genre into its own row
exploded_movie_genres = filtered_movies.explode('Genre')
exploded_tv_show_genres = filtered_tv_shows.explode('Show Genre')

# Step 3: Group by 'Genre' and count the number of occurrences
# For movies
movie_genre_counts = exploded_movie_genres.groupby('Genre').size().reset_index(name='Genre Count')

# For TV shows
tv_show_genre_counts = exploded_tv_show_genres.groupby('Show Genre').size().reset_index(name='Show Genre Count')

# Step 4: Merge the movie and TV show genre counts
combined_genre_counts = pd.merge(movie_genre_counts, tv_show_genre_counts, left_on='Genre', right_on='Show Genre', how='outer').fillna(0)

# Convert counts to integer type
combined_genre_counts['Genre Count'] = combined_genre_counts['Genre Count'].astype(int)
combined_genre_counts['Show Genre Count'] = combined_genre_counts['Show Genre Count'].astype(int)

# Step 5: Calculate total count and get top 5 genres
combined_genre_counts['Total Count'] = combined_genre_counts['Genre Count'] + combined_genre_counts['Show Genre Count']
top_5_genres = combined_genre_counts.sort_values(by='Total Count', ascending=False).head(5)

# Display the top 5 genres
print("Top 5 Genres for Movies and TV Shows:")
print(top_5_genres[['Genre', 'Total Count']])
````

## **Finding Predominant Streaming Service**

````python
# Write Your Code Here
# Step 1: Count occurrences of each streaming service for movies
movie_service_counts = movie_df.groupby('Streaming').size().reset_index(name='Movie Count')

# Step 2: Count occurrences of each streaming service for TV shows
tv_show_service_counts = tv_show_df.groupby('Show Streaming').size().reset_index(name='TV Show Count')

# Step 3: Merge the movie and TV show service counts into a combined DataFrame
combined_service_counts = pd.merge(
    movie_service_counts,
    tv_show_service_counts,
    left_on='Streaming',
    right_on='Show Streaming',
    how='outer'
).fillna(0)  # Fill NaN with 0

# Convert counts to integer type
combined_service_counts['Movie Count'] = combined_service_counts['Movie Count'].astype(int)
combined_service_counts['TV Show Count'] = combined_service_counts['TV Show Count'].astype(int)

# Step 4: Reset index and clean up the column names
combined_service_counts.reset_index(drop=True, inplace=True)

# Step 5: Calculate total count and get streaming services
combined_service_counts['Total Count'] = combined_service_counts['Movie Count'] + combined_service_counts['TV Show Count'] # Changed combined_genre_counts to combined_service_counts
top_5_genres = combined_service_counts.sort_values(by='Total Count', ascending=False) # Changed combined_genre_counts to combined_service_counts

# Display the streaming services
print(top_5_genres[['Streaming', 'Total Count']]) # Changed 'Streaming services' to 'Streaming'

````
